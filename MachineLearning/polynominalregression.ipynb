```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=6*np.random.rand(100,1)-3
y=0.5*x**2+1.5*x+2+np.random.randn(100,1)
plt.scatter(x,y,color='g')
plt.xlabel('X dataset')
plt.ylabel('Y dataset')
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=.25,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

regression_1 = LinearRegression()
regression_1.fit(X_train, Y_train)
Y_pred = regression_1.predict(X_test)
```

```python
from sklearn.metrics import r2_score
Y_pred = regression_1.predict(X_test)
score = r2_score(Y_test,Y_pred)
score
```

```python
## Letus visulaise the data
plt.plot(X_train,regression_1.predict(X_train),color='r')
plt.scatter(X_train,Y_train)
plt.xlabel('X Dataset')
plt.ylabel('Y Dataset')
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Assuming X_train_poly, X_test_poly, Y_train, and Y_test are already defined
# Define and fit the PolynomialFeatures transformation
poly = PolynomialFeatures(degree=2)  # Specify the degree as needed
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Train the Linear Regression model
regression_1 = LinearRegression()
regression_1.fit(X_train_poly, Y_train)

# Make predictions on the test set
Y_pred = regression_1.predict(X_test_poly)

# Evaluate the model
score = r2_score(Y_test, Y_pred)
print("R^2 Score:", score)

# Predict for the new dataset
X_new = np.linspace(-3, 3, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
Y_new_pred = regression_1.predict(X_new_poly)

# Optionally, plot the predictions
import matplotlib.pyplot as plt
plt.scatter(X_train, Y_train, color='blue', label='Training data')
plt.plot(X_new, Y_new_pred, color='red', label='Predictions')
plt.legend()
plt.show()

```
